```python
import cv2
import os
import time

# -------- CONFIGURATION --------
GESTURES = ["idle", "jump", "duck", "left", "right"]  # You can add more
FRAMES_PER_GESTURE = 30
SAVE_PATH = "gesture_data"
CAMERA_INDEX = 0  # change to 1 if external camera

# --------------------------------
def create_dirs():
    if not os.path.exists(SAVE_PATH):
        os.makedirs(SAVE_PATH)
    for gesture in GESTURES:
        gesture_path = os.path.join(SAVE_PATH, gesture)
        if not os.path.exists(gesture_path):
            os.makedirs(gesture_path)

def capture_gesture(gesture_name, frame_count=30):
    cap = cv2.VideoCapture(CAMERA_INDEX)
    if not cap.isOpened():
        print("Error: Could not access webcam.")
        return

    print(f"\nGet ready to record gesture: {gesture_name.upper()}")
    time.sleep(3)
    print("Recording started...")

    gesture_dir = os.path.join(SAVE_PATH, gesture_name)
    captured = 0

    while captured < frame_count:
        ret, frame = cap.read()
        if not ret:
            print("Error: Frame not captured properly.")
            break

        # Flip the frame (mirror effect)
        frame = cv2.flip(frame, 1)

        # Display the frame
        cv2.imshow(f"Recording: {gesture_name}", frame)

        # Save frame
        img_name = os.path.join(gesture_dir, f"{gesture_name}_{captured+1:03d}.jpg")
        cv2.imwrite(img_name, frame)
        captured += 1

        # Display progress
        print(f"Captured {captured}/{frame_count}", end='\r')

        # Wait between captures (slightly slower to allow repositioning)
        cv2.waitKey(100)

    print(f"\n{frame_count} frames captured for '{gesture_name}'.")
    cap.release()
    cv2.destroyAllWindows()

def main():
    create_dirs()
    print("Press 's' to start capturing next gesture or 'q' to quit.\n")

    for gesture in GESTURES:
        key = input(f"Ready to record '{gesture}' gesture? (press Enter to continue or 'q' to quit): ")
        if key.lower() == 'q':
            break
        capture_gesture(gesture, FRAMES_PER_GESTURE)

    print("\nAll gestures captured successfully!")

if __name__ == "__main__":
    main()
```
